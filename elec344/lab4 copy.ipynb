```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import scipy.interpolate as interp
import numpy as np
```

```python
# Set up data
S = [0,180,360,540,720,900,1080,1260,1350,1440,1620,1800]
T = [0.2205,0.2402,0.2637,0.2923,0.327,0.3695,0.4205,0.4764,0.499,0.5138,0.4346, 0]
f = interp.Akima1DInterpolator(S,T)
xnew = np.arange(0,1800, 0.1)
ynew = f(xnew)
S = [0,180,360,540,720,900,1080,1260,1440,1620,1800]
T = [0.2205,0.2402,0.2637,0.2923,0.327,0.3695,0.4205,0.4764,0.5138,0.4346, 0]
```

```python
# Full plot
figure(figsize=(12,8))
plt.plot(S, T, 'o', label="Simulation")
plt.plot(xnew, ynew, '-', label="theoretical")

plt.plot(1373, 0.52, '^')
plt.plot(1411,0.508, '^')
plt.annotate("Table 6", (1373, 0.52), textcoords="offset points", xytext=(-45,0))
plt.annotate("Table 7", (1411, 0.518), textcoords="offset points", xytext=(5,-30))
plt.xlim([0,1800])
plt.ylim([0,0.6])
plt.xlabel("Speed [rpm]")
plt.ylabel("Torque[N*m]")
plt.title("Torque vs Speed (Torque Speed Characteristic)")

plt.legend()
plt.savefig("lab4-part4c-1.png")
plt.show()
```
