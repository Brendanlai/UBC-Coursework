### Preprocessing the Training Set


```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('/content/gdrive/My Drive/Colab Notebooks/MANU465/Dataset/Train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

### Data Preprocessing

### Preprocessing the Test Set

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('/content/gdrive/My Drive/Colab Notebooks/MANU465/Dataset/Test',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
Model = tf.keras.models.Sequential()

Model.add(tf.keras.layers.Conv2D(filters=3, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
Model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
Model.add(tf.keras.layers.Conv2D(filters=3, kernel_size=3, activation='relu'))
```

```python
Model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
Model.add(tf.keras.layers.Flatten())
```

```python
Model.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
Model.add(tf.keras.layers.Dense(units=6, activation='softmax')) # units = number of classes
```

```python
Model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
Model.fit(x = training_set, validation_data = test_set, epochs = 25) #check with 20, 25 and 30
```

```python
Model.evaluate(training_set)
```

```python
Model.evaluate(test_set)
```
