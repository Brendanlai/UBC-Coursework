```python
# Use data from the last 12 months
R1 = [1,1.003,1.005,1.007,1.002,1.001,1.005,1.004,1.004,1.008,1.007,1.002]
R2 = [1.044,1.015,1.024,1.027,1.040,0.995,1.044,1.060,1,1.03,0.963,1.005]
R3 = [1.068,1.051,1.062,0.98,0.991,0.969,1.086,1.043,0.963,0.949,1.034,1.022]
R4 = [1.016,1.039,0.994,0.971,1.009,1.030,1.007,1.023,1.040,1.012,1.023,0.995]
R5 = [1.035,1.046,1.0008,0.989,1.021,0.997,1.024,1.028,1.038,1.011,0.943,0.999]
R6 = [1.032,1.047,1.01,0.973,1.02,0.989,1.028,1.04,1.04,1.07,0.974,0.995]
R7 = [1.004,1.028,1.021,0.985,1.02,1.02,0.991,1.018,0.999,1.039,1.016,1.018]
R8 = [0.987,1.049,1.036,1.053,0.996,0.999,1.026,1.053,0.985,1.028,1.048,1.023]
R9 = [1.014,1.073,1.002,0.977,1.03,1.007,0.999,1.003,1.015,1.029,1.055,1]
T = len(R1)

# calculate the r_j's
r1 = 1/T * np.sum(R1)
r2 = 1/T * np.sum(R2)
r3 = 1/T * np.sum(R3)
r4 = 1/T * np.sum(R4)
r5 = 1/T * np.sum(R5)
r6 = 1/T * np.sum(R6)
r7 = 1/T * np.sum(R7)
r8 = 1/T * np.sum(R8)
r9 = 1/T * np.sum(R9)

#Risk coeff for the objectives
y1=y2=y3=y4=y5=y6=y7=y8=y9= 0

for i in range (0,len(R1)):
    y1 += abs(R1[i] - r1)
    y2 += abs(R2[i] - r2)
    y3 += abs(R3[i] - r3)
    y4 += abs(R4[i] - r4)
    y5 += abs(R5[i] - r5)
    y6 += abs(R6[i] - r6)
    y7 += abs(R7[i] - r7)
    y8 += abs(R8[i] - r8)
    y9 += abs(R9[i] - r9)
```

```python
# for mu = 10^i (i = -2, -1, 0, 1, 2)
for i in range (-2, 3):
    mu = 10**i
    # instantiate problem
    optimum = pulp.LpProblem("max",pulp.LpMaximize)
    
    # investing optios
    x1 = pulp.LpVariable("x1", lowBound=0) # SHY
    x2 = pulp.LpVariable("x2", lowBound=0) # XLB
    x3 = pulp.LpVariable("x3", lowBound=0) # XLE
    x4 = pulp.LpVariable("x4", lowBound=0) # XLF
    x5 = pulp.LpVariable("x5", lowBound=0) # XLI
    x6 = pulp.LpVariable("x6", lowBound=0) # XLK
    x7 = pulp.LpVariable("x7", lowBound=0) # XLP
    x8 = pulp.LpVariable("x8", lowBound=0) # XLU
    x9 = pulp.LpVariable("x9", lowBound=0) # XLV
    
    # Constraints
    optimum += x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9 == 1 
    #objective function
    optimum += mu*(r1*x1+r2*x2+r3*x3+r4*x4+r5*x5+r6*x6+r7*x7+r8*x8+r9*x9) - 1/T*(x1*y1+x2*y2+x3*y3+x4*y4+x5*y5+x6*y6+x7*y7+x8*y8+x9*y9),"objective function"

    optimum.solve()
    print("decision variable values for mu = ", mu)
    for var in optimum.variables():
        print(var.name,"=",var.varValue)
    print("max: ", pulp.value(optimum.objective), "\n")
```
