```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
# Initiate knowns
Ra = 0.3
La = 5*10**(-3)
Ke = 0.4
kT = 0.4
Jm = 0.025
t_rated = 5
```

```python
#Va = 120V
Va = 120
x1 = np.arange(0,500,0.1)
y1 = (kT / Ra)*(Va - 0.4 * x1)

Va = 75
x2 = np.arange(0,500,0.1)
y2 = (kT / Ra)*(Va - 0.4 * x2)

Va = 45
x3 = np.arange(0,500,0.1)
y3 = (kT / Ra)*(Va - 0.4 * x3)

figure(figsize=(12,8))
plt.plot(x1, y1, label="120V")
plt.plot(x2, y2, label="75V")
plt.plot(x3, y3, label="75V")
plt.xlim([0,350])
plt.ylim([0,175])
plt.xlabel("Speed [rpm]")
plt.ylabel("Torque[N*m]")
plt.title("Torque vs Speed (Torque Speed Characteristic)")
plt.legend()
```
